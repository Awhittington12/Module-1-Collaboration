```python
Small = False
Green = True
if Small and Green == True:
    print("Pea")
elif Small == True:
    print("Pea, Cherry")
elif Green == True:
    print("Pea, Watermelon")
else:
    print("Pumpkin")
```

```python
guess_me = 7
number = 3
while number != guess_me:
    if number <= guess_me:
        print(number)
        print("Too Low")
        number += 1
    elif number >= guess_me:
        print("oops!")
        break
else:
    print(number)
    print("Found it!")
```

```python
guess_me = 5
for number in range(10):
    if number < guess_me:
        print(number)
        print("Too low")
        number += 1
    elif number == guess_me:
        print(number)
        print("Found it!")
        break
else:
    print(number)
    print("Oops")
```
